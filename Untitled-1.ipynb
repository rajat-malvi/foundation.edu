```python
from urllib.parse import urlparse, parse_qs

def extract_youtube_id(url):
    # For short youtu.be links
    if "youtu.be" in url:
        return url.split('/')[-1]
    # For regular youtube.com/watch?v= links
    if "youtube.com" in url:
        query = urlparse(url).query
        params = parse_qs(query)
        return params.get('v', [None])[0]
    return None

data = []
id_counter = 1

for card in syllabus:
    topic = card.get('topic', 'No Title')
    references = card.get('references', [])
    codes_and_datasets = card.get('codesAndDatasets', [])
    
    for link in card.get('youTubeLinks', []):
        video_id = extract_youtube_id(link['url'])
        thumbnail_url = f"https://img.youtube.com/vi/{video_id}/hqdefault.jpg" if video_id else ""

        entry = {
            "id": id_counter,
            "title": f"{topic} - {link['title']}",
            "status": "not-started",
            "youtubeLink": link['url'],
            "reference": references[0]['url'] if references else "",
            "code": "",
            "dataset": "",
            "imageUrl": thumbnail_url
        }

        # Assign first valid code and dataset URLs (similar to before)
        for item in codes_and_datasets:
            if not entry["code"] and "github" in item["url"].lower():
                entry["code"] = item["url"]
            elif not entry["code"] and "code" in item["title"].lower():
                entry["code"] = item["url"]

            if not entry["dataset"] and (
                "kaggle" in item["url"].lower() or "dataset" in item["title"].lower() or "uci" in item["url"].lower()
            ):
                entry["dataset"] = item["url"]

        data.append(entry)
        id_counter += 1
```
